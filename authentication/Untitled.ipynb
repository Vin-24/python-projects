```python
def register():
    with open("database.txt", "r") as db:
        username = input("Create username: ")
        password = input("Create password: ")
        password1 = input("Confirm password: ")
        
        d = []
        f = []
        for i in db:
            a, b = i.split(",")
            a = a.strip()
            b = b.strip()
            d.append(a)
            f.append(b)
        
        data = dict(zip(d, f))
        #print(data)
        
        if password != password1: 
            print("Passwords don't match. Restarting...")
            register()
        else:
            if len(password) <= 6:
                print("Password too short. Restarting...")
                register()
            elif username in d:
                print("Username already exists. Restarting...")
                register()
            else:
                with open("database.txt", "a") as db:
                    db.write(username + "," + password + "\n")
                print("Registration successful!")

```

```python
def access():
    with open("database.txt", "r") as db:
        username = input("Enter your username:")
        password = input("Enter your password:")

        if len(username) > 0 and len(password) > 0:
            d = []
            f = []
            for i in db:
                a, b = i.split(",")
                b = b.strip()
                d.append(a)
                f.append(b)
            data = dict(zip(d, f))
            #print(data)
            try:
                if data[username]:
                    try:
                        if password == data[username]:
                            print("Login success")
                            print("Hi, ", username)
                        else:
                            print("Password or username incorrect")
                    except:
                        print("Incorrect password or username")
                else:
                    print("Username or password doesn't exist")
            except:
                print("Username or password doesn't exist")
        else:
            print("Please enter the correct data")

```

```python
def reset_password_for_username():
    username=input("enter username")
    with open("database.txt", "r") as db:
        d = []
        f = []
        for i in db:
            a, b = i.split(",")
            a = a.strip()
            b = b.strip()
            d.append(a)
            f.append(b)

    data = dict(zip(d, f))

    if username in data:
        new_password = input("Enter a new password: ")
        new_password1=input("confirm new password")
        if new_password!=new_password1:
            print("Passwords don't match. Restarting...")
            reset_password_for_username()
        else:
            data[username] = new_password

            with open("database.txt", "w") as db:
                for user, password in data.items():
                    db.write(user + "," + password + "\n")

            print("Password reset successful")
            access()  
    else:
        print("Username not found")
        access()      
```
